```python
from random import randint
```

```python
def print_tries_number(n):
    last_digit = n % 10
    if last_digit == 1:
        print('\nУ вас осталась {} попытка.'.format(n))
    elif last_digit in range(2, 5):
        print('\nУ вас осталось {} попытки.'.format(n))
    elif last_digit in range(5, 10):
        print('\nУ вас осталось {} попыток.'.format(n))
    else:
        print('\nУ вас осталось {} попыток.'.format(n))
    return
```

```python
def load_words_by_theme(theme):
    with open('{}.txt'.format(theme), 'r') as file:
        words = file.read().split()
    return words
```

```python
def draw_hanged(number_of_trials):
    n = 6 - number_of_trials
    if n == 0:
        print('          ')
        print('          ')
        print('          ')
        print('          ')
        print('          ')
        print('__________')
    elif n == 1:
        print('   __     ')
        print('   ||     ')
        print('   ||     ')
        print('   ||     ')
        print('   ||     ')
        print('__/__\____')
    elif n == 2:
        print('   _____  ')
        print('   ||  |  ')
        print('   ||     ')
        print('   ||     ')
        print('   ||     ')
        print('__/__\____')
    elif n == 3:
        print('   _____  ')
        print('   ||  |  ')
        print('   ||  o  ')
        print('   ||     ')
        print('   ||     ')
        print('__/__\____')
    elif n == 4:
        print('   _____  ')
        print('   ||  |  ')
        print('   ||  o  ')
        print('   ||  |  ')
        print('   ||     ')
        print('__/__\____')
    elif n == 5:
        print('   _____  ')
        print('   ||  |  ')
        print('   ||  o  ')
        print('   || /|\ ')
        print('   ||     ')
        print('__/__\____')
    elif n == 6:
        print('   _____  ')
        print('   ||  |  ')
        print('   ||  o  ')
        print('   || /|\ ')
        print('   || / \ ')
        print('__/__\____')
```

```python
def print_word(word, guessed):
    if len(word) != len(guessed):
        return
    print()
    for i in range(len(word)):
        if guessed[i]:
            print(word[i].upper(), end=' ')
        else:
            print('_', end=' ')
    print()
    print('\nВведите букву:\n')
```

```python
def check_guess(letter, word, guessed, tried):
    word = word.lower()
    if not letter.isalpha() and len(letter) == 1:
        print('Вы ввели не букву.')
        return guessed, False
    
    letter = letter.lower()
    if letter in tried:
        print('Вы уже называли эту букву!')
        return guessed, False
    
    found = []
    
    if letter in word:
        for i in range(len(word)):
            if word[i] == letter:
                found.append(i)
    
    if len(found) == 0:
        print('Не угадали :(')
        return guessed, True

    print('Буква "{}" отгадана!'.format(letter))
    for i in found:
        guessed[i] = True
    
    return guessed, False
```

```python
def print_menu():
    print('\nВыберите тему слов для игры (введите число):')
    print('1. Танцы.')
    print('2. Породы кошек.')
    print('3. Фамилии актёров.\n')
```

```python
def main():
    while True:
        print_menu()
        theme_num = int(input())
        if (theme_num not in range(1, 4)):
            print('Неверный ввод. Введите число от 1 до 3.\n')
            continue

        theme = ''
        if theme_num == 1:
            theme = 'dances'
            print('Выбрана тема: "Танцы"!\n')
        if theme_num == 2:
            theme = 'cats'
            print('Выбрана тема: "Породы кошек"!\n')
        if theme_num == 3:
            theme = 'actors'
            print('Выбрана тема: "Фамилии актёров"!\n')

        words = load_words_by_theme(theme)
        
        word = words[randint(1, len(words)) - 1]
        guessed = [0] * len(word)
        tries_number = 6
        tried = []
        word_guessed = False
        failed = False
        while not word_guessed and not failed:
            print_tries_number(tries_number)
            draw_hanged(tries_number)
            print_word(word, guessed)
            letter = input()
            
            use_trie = False
            guessed, use_trie = check_guess(letter, word, guessed, tried)
            if not letter in tried:
                tried.append(letter)
            if use_trie:
                tries_number -= 1
            if all(guessed):
                word_guessed = True
            elif tries_number == 0:
                draw_hanged(tries_number)
                print('\nСожалеем, но вас повесили. Возможно, вашей реинкарнации удастся выиграть?\n')
                contin = True
                while contin:
                    print('Введите 1, чтобы продолжить и 0, чтобы выйти.')
                    end_num = int(input())
                    if end_num not in range(2):
                        print('Неверный ввод.')
                        continue
                    if end_num == 0:
                        return
                    else:
                        contin = False
                        failed = True
        if word_guessed:
            draw_hanged(tries_number)
            contin = True
            print('\nПоздравляем, вы отгадали слово! Вы справляетесь, попробуйте ещё!\n')
            contin = True
            while contin:
                print('Введите 1, чтобы продолжить и 0, чтобы выйти.')
                end_num = int(input())
                if end_num not in range(2):
                    print('Неверный ввод.')
                    continue
                if end_num == 0:
                    return
                else:
                    contin = False
```

```python
if __name__ == "__main__":
    main()
```
